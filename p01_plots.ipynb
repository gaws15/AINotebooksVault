```python
import matplotlib.pyplot as plt
import numpy as np

# Ideal values
ideal_values = [3, 6, 11, 16, 21, 26, 31, 36, 41, 46, 51, 56]

# Actual data for each sample (replace with your actual data)
actual_data_samples = [
    [3, 7, 11, 15, 20, 24, 29, 34, 38, 43, 48, 53],  # Sample 1
    [3, 6, 10, 15, 19, 24, 29, 33, 38, 43, 48, 53],  # Sample 2
    [3, 6, 10, 15, 19, 24, 29, 34, 38, 43, 48, 53],  # Sample 3
    [3, 7, 11, 15, 19, 24, 29, 33, 38, 43, 48, 53],  # Sample 4
    [3, 6, 10, 15, 20, 24, 29, 34, 38, 43, 48, 53]   # Sample 5
]

# Convert actual_data_samples to a NumPy array
actual_data_samples = np.array(actual_data_samples)

# Calculate the average of all samples
average_sample = np.mean(actual_data_samples, axis=0)

# Combine all samples and ideal values for the bar chart
all_samples = actual_data_samples.tolist() + [average_sample.tolist()]  # Convert to lists for plotting
sample_labels = [f'Sample {i + 1}' for i in range(len(actual_data_samples))] + ['Average']

# Create the bar chart
bar_width = 0.15
x_positions = np.arange(len(ideal_values))

fig, ax = plt.subplots(figsize=(12, 6))

for i, sample in enumerate(all_samples):
    ax.bar(x_positions + i * bar_width, sample, width=bar_width, label=sample_labels[i])

# Plot ideal values as a line with markers
ax.plot(x_positions + (len(all_samples) / 2 - 0.5) * bar_width, ideal_values,
        marker='o', linestyle='--', color='red', label='CNC Values(z-axis)')

# Set x-axis labels and title
ax.set_xticks(x_positions + (len(all_samples) / 2 - 0.5) * bar_width)
ax.set_xticklabels([f'Layer {i + 1}' for i in range(len(ideal_values))])
ax.set_ylabel('Bead Height (mm)')
#ax.set_title('Comparison of Samples and Ideal Values')

# Add legend and grid
ax.legend(loc='best', fontsize='small')
ax.grid(True)

# Adjust layout to prevent legend from overlapping
plt.tight_layout()

plt.show() # Or
plt.savefig('CNC VS MEASURED.png')
```
