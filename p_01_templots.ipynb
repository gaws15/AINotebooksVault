```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
import seaborn as sns  # Import seaborn
# Generate sample data

# Generate sample data
def generate_data():
    time_points = np.array([25, 75, 125, 175, 225, 275, 325, 375, 425, 475])
    n_measurements = 10
    n_timepoints = len(time_points)

    # Reshape heights and temperatures to be 2D arrays
    heights = np.array([5,4.6667,	4.33,	4	,4,4,4,3.6667,	3.6667,	3.6667]).reshape(-1, 1)
    temperatures = np.array([115, 114, 102, 98, 92, 91, 86, 82, 80, 77]).reshape(-1, 1)

    return time_points, heights, temperatures

def create_monitoring_plot(time_points, heights, temperatures):

    sns.set_style('whitegrid')  # Use seaborn to set the style if desired
    fig = plt.figure(figsize=(15, 10))
    gs = GridSpec(2, 1, height_ratios=[2, 2], hspace=0.3)

    # Height plot
    ax1 = fig.add_subplot(gs[0])
    ax2 = fig.add_subplot(gs[1])
    # Plot individual measurements as light blue dots
    for i in range(heights.shape[1]):
        ax1.scatter(time_points, heights[:, i], color='lightblue', alpha=0.3, s=30)

    # Plot mean height as a blue line
    mean_height = np.mean(heights, axis=1)
    ax1.plot(time_points, mean_height, color='blue', linewidth=2, label='Mean Height')

    # Add standard deviation area
    std_height = np.std(heights, axis=1)
    ax1.fill_between(time_points,
                     mean_height - std_height,
                     mean_height + std_height,
                     color='blue', alpha=0.2, label='1 STD')

    ax1.set_xlabel('Time (seconds)')
    ax1.set_ylabel('Layer Height (mm)')
    ax1.set_title('Layer Height Variation Over Time')
    ax1.grid(True, linestyle='--', alpha=0.7)
    ax1.legend()

    # Temperature plot
    ax2 = fig.add_subplot(gs[1])
    ax2.axhline(y=105, color='green', linestyle='--', label='tg=105')
    ax2.legend() # Update the legend to include the new line

    # Plot individual measurements as light red dots
    for i in range(temperatures.shape[1]):
        ax2.scatter(time_points, temperatures[:, i], color='mistyrose', alpha=0.3, s=30)

    # Plot mean temperature as a red line
    mean_temp = np.mean(temperatures, axis=1)
    ax2.plot(time_points, mean_temp, color='red', linewidth=2, label='Mean Temperature')

    # Add standard deviation area
    std_temp = np.std(temperatures, axis=1)
    ax2.fill_between(time_points,
                     mean_temp - std_temp,
                     mean_temp + std_temp,
                     color='red', alpha=0.2, label='1 STD')

    ax2.set_xlabel('Time (seconds)')
    ax2.set_ylabel('Temperature (C)')
    ax2.set_title('Temperature Variation Over Time')
    ax2.grid(True, linestyle='--', alpha=0.7)
    ax2.legend()

    plt.suptitle('3D Print Layer Monitoring', fontsize=16, y=1.02)

    # Add text box with statistics
    stats_text = (f'Height Stats:\n'
                 f'Mean: {np.mean(heights):.2f}mm\n'
                 f'Std: {np.std(heights):.2f}mm\n\n'
                 f'Temperature Stats:\n'
                 f'Mean: {np.mean(temperatures):.2f}C\n'
                 f'Std: {np.std(temperatures):.2f}C')

    plt.figtext(1.02, 0.5, stats_text, bbox=dict(facecolor='white', alpha=0.8),
                fontsize=10, ha='left', va='center')

    plt.tight_layout()
    return fig

# Generate and plot data
time_points, heights, temperatures = generate_data()
fig = create_monitoring_plot(time_points, heights, temperatures)

# Save the plot
plt.savefig('layer_monitoring.png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
time_points = np.array([125, 175, 225, 275, 325, 375, 425, 475, 525, 575])

# Temperature Data (with leading Nones for alignment)
T3_prime = [120, 116, 110, 109, 104, 100, 98, 95, 93, 91]          # Starts at 125s
T4_prime = [None, 123, 117, 116, 111, 107, 105, 102, 100, 98]       # Starts at 175s
T5_prime = [None, None, 123, 122, 117, 113, 111, 108, 106, 104]     # Starts at 225s

# Height Data (with leading Nones for alignment)
L3 = [5, 4.6667, 4.33, 4, 4, 4, 4, 3.6667, 3.6667, 3.6667]
L4 = [None, 4.667, 4.3333, 4.3333, 4.333, 4.333, 4.333, 4.3333, 4, 4]
L5 = [None, None, 5, 4.6667, 4.667, 4.6667 , 4.6667,4.333, 4.333, 4.3333]

# Function to align data with time_points
def process_layer(temp, height, start_idx):
    # Remove Nones and trim to valid length
    temp_clean = [v for v in temp if v is not None]
    height_clean = [v for v in height if v is not None]
    min_length = min(len(temp_clean), len(height_clean))
    # Get time points for this layer
    time = time_points[start_idx : start_idx + min_length]
    return time, temp_clean[:min_length], height_clean[:min_length]

# Create subplots
fig, axes = plt.subplots(3, 1, figsize=(12, 10), sharex=True)
tg_value = 105  # Glass transition temperature

# Layer configurations
layers = [
    {'temp': T3_prime, 'height': L3, 'start_idx': 0, 'name': 'Layer 3 (Starts at 125s)'},
    {'temp': T4_prime, 'height': L4, 'start_idx': 1, 'name': 'Layer 4 (Starts at 175s)'},
    {'temp': T5_prime, 'height': L5, 'start_idx': 2, 'name': 'Layer 5 (Starts at 225s)'}
]

for idx, layer in enumerate(layers):
    ax = axes[idx]
    time, temp, height = process_layer(layer['temp'], layer['height'], layer['start_idx'])

    # Plot Temperature (Left Y-axis)
    ax.plot(time, temp, 'r-o', markersize=8, label='Temperature')
    ax.axhline(tg_value, color='green', linestyle='--', linewidth=1, label=f'Tg ({tg_value}C)')
    ax.set_ylabel('Temperature (C)', color='r', fontsize=12)
    ax.tick_params(axis='y', labelcolor='r')
    ax.grid(True, linestyle=':', alpha=0.7)
    ax.set_title(layer['name'], fontsize=12, pad=10)

    # Plot Height (Right Y-axis)
    ax2 = ax.twinx()
    ax2.plot(time, height, 'b--s', markersize=8, label='Height')
    ax2.set_ylabel('Height', color='b', fontsize=12)
    ax2.tick_params(axis='y', labelcolor='b')

    # Combine legends
    lines1, labels1 = ax.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax.legend(lines1 + lines2, labels1 + labels2, loc='upper right')

# Final adjustments
plt.xlabel('Time (s)', fontsize=12)
plt.xlim(100, 600)  # Extend x-axis slightly for clarity
plt.suptitle('Temperature and Height Trends Across Layers', fontsize=14, y=0.98)
plt.tight_layout()
plt.savefig('temperature_height_layers.png', dpi=300, bbox_inches='tight')
plt.show()

```
